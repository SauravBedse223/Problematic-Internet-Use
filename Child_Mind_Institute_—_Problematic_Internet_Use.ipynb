## Step 5: Identify and Handle Outliers

### Method 1: Using the IQR (For Skewed Data)

## Outlier Detection

```python
import numpy as np

# Define a function to identify outliers
def identify_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    # Outliers are values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR
    outliers = df[(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))]
    return outliers

# Apply the function to numeric columns and get summary of outliers
numeric_cols = df_train.select_dtypes(include=['float64', 'int64']).columns
print("Outliers Detection :")
for col in numeric_cols:
    outliers = identify_outliers(df_train, col)
    print(f"{col}: {len(outliers)} outliers")
```

```python
# Function to handle outliers by imputing them with the median
def impute_outliers_with_median(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Impute outliers with the median
    median_value = df[col].median()
    df.loc[df[col] < lower_bound, col] = median_value
    df.loc[df[col] > upper_bound, col] = median_value
```

```python
# Apply the function to all numeric columns
numeric_cols = df_train.select_dtypes(include=['float64', 'int64']).columns

print("Handling Outliers:")
for col in numeric_cols:
    outliers = identify_outliers(df_train, col)
    print(f"{col}: {len(outliers)} outliers identified and imputed")
    impute_outliers_with_median(df_train, col)
```

## Implementing Transformations

```python
import numpy as np
import pandas as pd
from scipy.stats import boxcox


# Function to apply transformations based on skewness type for training data
def transform_skewed_features_train(df, skew_threshold=0.5):
    for column in df_train.columns:
        if df_train[column].skew() > skew_threshold:  # Positively skewed
            df_train[column] = np.log1p(df_train[column])  # Log transformation
        elif df_train[column].skew() < -skew_threshold:  # Negatively skewed
            df_train[column] = np.square(df_train[column])  # Square transformation
    return df_train


# Function to apply transformations based on skewness type for testing data
def transform_skewed_features_test(df, skew_threshold=0.5):
    for column in df_test.columns:
        if df_test[column].skew() > skew_threshold:  # Positively skewed
            df_test[column] = np.log1p(df_test[column])  # Log transformation
        elif df_test[column].skew() < -skew_threshold:  # Negatively skewed
            df_test[column] = np.square(df_test[column])  # Square transformation
    return df_test


# Transform your data
df_transformed_train = transform_skewed_features_train(df_train)
df_transformed_test = transform_skewed_features_test(df_test)
```
