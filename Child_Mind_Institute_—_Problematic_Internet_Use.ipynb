## Step 2: Handling Missing Values

```python
# Impute missing numerical values in df_train
df_train['Physical-BMI'].fillna(df_train['Physical-BMI'].median(), inplace=True)
df_train['Physical-Height'].fillna(df_train['Physical-Height'].median(), inplace=True)
df_train['Physical-Weight'].fillna(df_train['Physical-Weight'].median(), inplace=True)
df_train['CGAS-CGAS_Score'].fillna(df_train['CGAS-CGAS_Score'].mean(), inplace=True)
df_train['PCIAT-PCIAT_Total'].fillna(df_train['PCIAT-PCIAT_Total'].mean(), inplace=True)

# Impute missing categorical values with mode (most frequent value)
df_train['Basic_Demos-Enroll_Season'].fillna(df_train['Basic_Demos-Enroll_Season'].mode()[0], inplace=True)
df_train['Basic_Demos-Sex'].fillna(df_train['Basic_Demos-Sex'].mode()[0], inplace=True)
df_train['Basic_Demos-Age'].fillna(df_train['Basic_Demos-Age'].mode()[0], inplace=True)
```

## Step : 3 Drop columns

```python
# Drop the columns PCIAT_1 to PCIAT_20
columns_to_drop1 = [f'PCIAT-PCIAT_0{i}' for i in range(1, 10)]
columns_to_drop2 = [f'PCIAT-PCIAT_{i}' for i in range(10, 21)]
df_train.drop(columns=columns_to_drop1, inplace=True)
df_train.drop(columns=columns_to_drop2, inplace=True)
```

## STEP 4 : Removing Null Values


```python
df_train.isnull().sum()
```

```python
df_train.head()
```

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Identify columns with missing values
cols_with_na = df_train.columns[df_train.isnull().any()]

# Filter to only include numeric columns with missing values
numeric_cols_with_na = df_train[cols_with_na].select_dtypes(include=['float64', 'int64']).columns
```

```python
# Plot histograms for columns with missing values
for col in numeric_cols_with_na:
    plt.figure(figsize=(6, 4))
    sns.histplot(df_train[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Now, fill missing values based on distribution
for col in numeric_cols_with_na:
    if df_train[col].skew() < 1:  # Symmetrical distribution
        df_train[col].fillna(df_train[col].mean(), inplace=True)

    else:  # Skewed distribution
        df_train[col].fillna(df_train[col].median(), inplace=True)
```

## Step 5 : One Hot Encoding


```python
from sklearn.preprocessing import LabelEncoder

# Identify categorical columns
categorical_cols = df_train.select_dtypes(include=['object']).columns

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Apply Label Encoding to each categorical column in both train and test sets
for col in categorical_cols:
    df_train[col] = label_encoder.fit_transform(df_train[col].astype(str))
```

```python
# Check the transformations
print("Label Encoded Columns in Train Dataset:")
df_train_categorical_cols = df_train[categorical_cols]
df_train_categorical_cols.head()
```
